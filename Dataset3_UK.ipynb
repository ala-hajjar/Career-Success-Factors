```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
df=pd.read_csv('salary_uk.csv')
df
```

```python
# To focus on our interest, a subset dataframe is created
df_sub=df[['Category', 'SalaryNormalized']]
```

```python
df_sub
```

```python
# Renaming columns
df_sub = df_sub[['Category', 'SalaryNormalized']].rename(columns={
    'Category': 'Job Category',
    'SalaryNormalized': 'Salary'})
```

```python
# checking if it worked
df_sub
```

```python
df_grouped = df_sub.groupby('Job Category')['Salary'].agg(['mean', 'median','min','std'])
df_grouped
```

```python
# To remove the word jobs from the end of each Job Category ( removing repitition)
df_grouped.index = df_grouped.index.str.replace(' Jobs', '', regex=False)
df_grouped
```

```python
# Creating a bar plot for mean salary by job category
plt.figure(figsize=(12, 6))
sns.barplot(x=df_grouped.index, y=df_grouped['mean'], palette="viridis")
sns.set_style("whitegrid")
plt.title('Mean Salary by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Mean Salary per year')
plt.xticks(rotation=90)
plt.show()
```

#It is crowded, so we will focus on 10 highest and lowests jobs, for a clear visualization

```python
#Creating a function to visualize 10 highest and 10 lowest salaries per job category/field
sns.set_style("whitegrid")

def plot_salary_by_category(df_grouped, subset='top', n=10):
    if subset == 'top':
        selected_categories = df_grouped.nlargest(n, 'mean')
        title = f'Top {n} Highest Paying Jobs'
    elif subset == 'bottom':
        selected_categories = df_grouped.nsmallest(n, 'mean')
        title = f'Bottom {n} Paying Jobs'
    else:
        raise ValueError("Invalid subset type. Use 'top' or 'bottom'.")


    plt.figure(figsize=(11, 5))
    ax = sns.barplot(x=selected_categories.index, y=selected_categories['mean'], palette="viridis")

    plt.title(title, fontsize=14)
    plt.xlabel('Job Category', fontsize=12)
    plt.ylabel('Mean Salary', fontsize=12, labelpad=15)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)


    for p in ax.patches:
        ax.annotate(f'{p.get_height():,.0f}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom', fontsize=10, color='black', weight='bold')

    plt.show()
```

```python
# Top 10 highest paying jobs
plot_salary_by_category(df_grouped, subset='top', n=10)
```

```python
# Bottom 10 lowest paying
plot_salary_by_category(df_grouped, subset='bottom', n=10)
```
